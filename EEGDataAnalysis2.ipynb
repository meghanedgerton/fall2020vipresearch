### Visualization

```python
samp1time = Sample1['Timestamp']
channel0 = Sample1['EXG Channel 0']
channel1 = Sample1['EXG Channel 1']
channel2 = Sample1['EXG Channel 2']
channel3 = Sample1['EXG Channel 3']

plt.figure()
plt.plot(samp1time,channel0)

plt.plot(samp1time,channel1)

plt.plot(samp1time,channel2)

plt.plot(samp1time,channel3)

plt.ylabel('Voltage Reading in mV')
plt.xlabel('Timestamp')
plt.title('Subject 1 Sample 1')
plt.legend(loc = 'upper right',bbox_to_anchor=(0.5, -0.15), ncol= 2)
```
